## Simple Linear Regression

## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2. Import the dataset

```python
df=pd.read_csv('50_Startups.csv')
```

```python
df
```

## Step 3 . Check the State Column

```python
df['State'].value_counts()
```

## Step 4 . Splitting dataset in to input and output

```python
x=df.iloc[::,:4:]
y=df.iloc[::,4::]
x
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Perform Transformation

```python
x1=pd.get_dummies(x,columns=['State'],drop_first=True)
```

```python
x1
```

## Step 6 . Dummy variable trap

```python
# Already Performed using | drop_first =True
```

## Step 7 Splitting dataset in to Train and Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y,train_size=0.8,random_state=42)
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

## Step 9  Fit the data

```python
print(model.fit(x_train,y_train))
```

## Step 10  Predict the data

```python
y_pred=model.predict(x_test)
y_pred
```

## Step 11  Display Result

```python
# y_test and y_predict
```

```python
plt.scatter(x_test['Administration'],y_test)
plt.scatter(x_test['Administration'],y_pred)
```

## RSS

```python
np.sum((y_test.values-y_pred)**2)
```

```python
len(y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test.values,y_pred)*len(y_pred)
```

```python
len(y_pred)
```

## R Square

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

## Now use Polynomial Regression on Position_Salaries dataset

```python
df2=pd.read_csv('Position_Salaries.csv')
df2
```

```python
plt.scatter(df2['Level'],df2['Salary'])
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Position vs Salary')
```

```python
x=df2.iloc[::,1:2:]
y=df2['Salary']
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x1=poly.fit_transform(x_train)
x1
```

```python
model=LinearRegression()
model.fit(x1,y_train)
```

```python
y_poly_predict=model.predict(poly.fit_transform(x_test))
y_poly_predict
```

```python
plt.scatter(df2['Level'],df2['Salary'])
plt.scatter(df2['Level'],model.predict(poly.fit_transform(x)))
```

```python
plt.scatter(x_test,y_test)
plt.scatter(x_test,y_poly_predict)
```

```python
np.sum((y_test.values-y_poly_predict)**2)
```
